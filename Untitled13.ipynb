```python
class C_rectangulo:
  def __init__(self,P_ancho,P_alto):
    self.at_ancho = P_ancho
    self.at_alto = P_alto
  
  def m_area(self):
    area = self.at_alto*self.at_ancho
    return area
  
  def m_perimetro(self):
    perimetro = (self.at_alto*2)+(self.at_ancho*2)
    return perimetro

r1 = C_rectangulo(4,2)
area = r1.m_area()
perimetro = r1.m_perimetro()
print("El area es: ", area)
print("El perimetro es: ", perimetro)

P_ancho = int(input("Ingrese el ancho del rectangulo: "))
P_alto = int(input("Ingrese el alto del rectangulo: "))

obj_rect2 = C_rectangulo(P_ancho, P_alto)
print("El area del rectangulo es: ",obj_rect2.m_area())
print("El perimetro del rectangulo es: ",obj_rect2.m_perimetro())
```

```python
class C_rectangulo:
  def __init__(self,P_ancho,P_alto):
    self.at_ancho = P_ancho
    self.at_alto = P_alto
  
  def m_area(self):
    area = self.at_alto*self.at_ancho
    return area
  
  def m_perimetro(self):
    perimetro = (self.at_alto*2)+(self.at_ancho*2)
    return perimetro
  def m_hipotenusa(self):
    Hipotenusa = (self.at_alto^2)+(self.at_ancho)
    return Hipotenusa

r1 = C_rectangulo(4,2)
area = r1.m_area()
perimetro = r1.m_perimetro()
Hipotenusa = r1.m_hipotenusa()
print("El area es: ", area)
print("El perimetro es: ", perimetro)
print("El valor de la hipotenusa es: ",Hipotenusa)
P_ancho = int(input("Ingrese el ancho del rectangulo: "))
P_alto = int(input("Ingrese el alto del rectangulo: "))

obj_rect2 = C_rectangulo(P_ancho, P_alto)
print("El area del rectangulo es: ",obj_rect2.m_area())
print("El perimetro del rectangulo es: ",obj_rect2.m_perimetro())
print("la hipotenusa del rectangulo es: ",obj_rect2.m_hipotenusa())
```
